```python
import pandas as pd
import numpy as np
from PIL import Image 
import os
# from tqdm import tqdm
from tqdm.notebook import tqdm
tqdm.pandas()
import csv

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
```

```python
# list all images names - done
# convert those list to 3d array df
# preprocess
# train it using cnn
# test it using Validation
# tune and check others solution

# female - 0
# male - 1
```

```python
listOfmaleNames = os.listdir("../data/male_faces")
listOffemaleNames = os.listdir("../data/female_faces")
```

```python
listOfmaleNames = ["/data/male_faces/"+name for name in listOfmaleNames]
```

```python
listOffemaleNames = ["/data/female_faces/"+name for name in listOffemaleNames]
```

```python
len(listOfmaleNames)
```

```python
len(listOffemaleNames)
```

```python
df_male = pd.DataFrame({"image_path":listOfmaleNames,"gender":1})
```

```python
df_female = pd.DataFrame({"image_path":listOffemaleNames,"gender":0})
```

```python
df = pd.concat([df_male,df_female],ignore_index=True)
```

```python
beforeImg = Image.open(".."+df["image_path"][233])
print(beforeImg.size)
beforeImg
```

```python
def image_to_array(path):
    newPath = "../"+path
    img = Image.open(newPath)
    compressedImg = img.convert('L')
    resizedImg = compressedImg.resize((50,50), Image.ANTIALIAS)
    normalizedImg =  np.array(resizedImg)/255.0
    flattenedImg = normalizedImg.reshape(-1)
    return flattenedImg
```

```python
tempImg = Image.open(".."+df["image_path"][233])
compressedImg = tempImg.convert('L')
resizedImg = compressedImg.resize((50,50), Image.ANTIALIAS)
print(resizedImg.size)
resizedImg
```

```python
df["image_array"] = df["image_path"].progress_apply(image_to_array)
```

```python
X = np.array(df["image_array"].tolist())
y = df["gender"]
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=41)
```

```python
model = LogisticRegression(solver='liblinear')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
joblib.dump(model,"LR_simple_model_1.pkl")
```

```python
def make_prediction_of_individual_img(path):
    array = image_to_array(path)
    prediciton = model.predict([array])
    if(prediciton[0]==1):
        return "Male"
    else:
        return "Female"
```
